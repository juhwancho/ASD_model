```python
import requests
import json

# Set up the API endpoint and parameters
url = 'https://reactome.org/ContentService/data/pathways/top/9606'
params = {
    'pageSize': 100,  # Maximum number of pathways per page
    'page': 1  # Start with the first page
}
combined={}
L1={}
L2={}
# Make a GET request to the API
response = requests.get(url, params=params)

# Check if the request was successful
if response.status_code == 200:
    # Parse the response data as JSON
    data = json.loads(response.text)
    
    # Create an empty dictionary to store the pathway info
    pathway_info = {}
    # Iterate over each pathway in the response data
    for pathway in data:
        # Get the pathway ID and name
        pathway_id = pathway['stId']
        pathway_name = pathway['name']
        L1[pathway_id]=pathway_name
        combined[pathway_id]=[]
        url2 = f'https://reactome.org/ContentService/data/pathway/{pathway_id}/containedEvents'
        response2 = requests.get(url2)
        if response2.status_code == 200:
            # Parse the response data as JSON
            data2 = json.loads(response2.text)
            for event in data2:
                if event['className']=='Pathway':
                    L2[event['stId']]=event['displayName']
                    combined[pathway_id].append(event['stId'])
                    
                   
        
     
       
else:
    print(f"Error: {response.status_code}")
```

```python
pathwaylist2=list(L1.keys())

genecnt=0
reactomegenedict2={}
pathwaycnt=0
for j in pathwaylist2:
    pathwaycnt+=1
    print(pathwaycnt)
    reactomegenedict2[j]=[]
    url=f'https://reactome.org/ContentService/data/participants/{j}/participatingPhysicalEntities'
    response = requests.get(url)
    if response.status_code == 200:
            data = json.loads(response.text)
            querylist=""
            querycnt=0
            for participant in data:
                if participant['className']=='Protein':
                    ids=participant['stId']
                    if querycnt!=0:
                        querylist+=", "
                    querylist+=ids
                    querycnt+=1
            url3=f'https://reactome.org/ContentService/data/query/ids'
            #jsonString = json.dumps(list(querylist))
            headers ={'Content-Type': 'text/plain'}
            response3 = requests.post(url=url3,headers=headers,data=querylist)
            if response3.status_code == 200:
                data2 = json.loads(response3.text)
                for res in data2:
                    if "geneName" in res["referenceEntity"].keys():
                        names=res["referenceEntity"]["geneName"][0]
                        reactomegenedict2[j].append(names)
                        genecnt+=1
print(genecnt)                        
                   
  
```
